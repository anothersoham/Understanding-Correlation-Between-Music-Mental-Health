```python
#Calling the specified rows from the dataset
cs1 = pd.read_csv('share-with-mental-and-substance-disorders.csv', skiprows = lambda x: x not in specific_rows)
```

```python
#Calling the dataset to check the revised structure
cs1
```

```python
#Renaming the columns for tidier data and precise understanding
cs1.rename(columns={"Prevalence - Mental disorders - Sex: Both - Age: Age-standardized (Percent)": "Mental Health Cases (Percentage)"}, inplace=True)
```

```python
#Calling the dataset to check the applied edit
cs1
```

```python
#Reindexing the columns
cs1.reindex(columns=['Entity', 'Code', 'Year', 'Mental Health Cases (Percentage)'])
```

```python
#Specifing the required columns for visualisation
cs2= cs1[['Year', 'Mental Health Cases (Percentage)']]
```

```python
#Calling the dataset to check the revised structure
cs2
```

```python
#Figure 9:Prevalence of Mental Health Disorders in US (1990-2019)
#Using the Seaborn to plot a linegraph of the percentage of mental illness between 1990 and 2019 in the US 

plt.figure(figsize=(12, 7), 
           dpi = 1200) 
 
ax1= sns.lineplot(data= cs2, x="Year", y= "Mental Health Cases (Percentage)", color="orange", lw=3.5)
ax1.set_title('Figure 9: Prevalence of Mental Health Disorders in the US (1990-2019)',fontsize=16)
plt.tick_params(axis = 'x', labelsize = 12)
plt.tick_params(axis = 'y', labelsize = 12)
plt.legend(labels=["The Percentage of Mental Health Cases"], title= "Data")
plt.show()
```
